## Lambda

### Ex1

```python
plus = lambda x,y:x + y
print(plus(1,2))
```

### Ex2

```python
#max of x, y
def mx(x,y):
  if x > y:
    return x
  else:
    return y

print(mx(1,3))
```

```python
mx = lambda x,y: x if x > y else y
print(mx(1,2))
```

## Map

### Ex1

```python
n = [4, 3, 2, 1]
#prints [16, 9, 4, 1]

def square (lst1):
  lst2 = []
  for num in lst1:
    lst2.append(num**2)
  return lst2

print(square(n))
```

```python
print(list(map(lambda x:x**2,n)))
```

```python
print([x**2 for x in n])
```

## Filter

```python
n = [4, 3, 2, 1]
#print[4,3]

def over_two(lst1):
  lst2 = [x for x in lst1 if x > 2]
  return lst2

print(over_two(n))
```

```python
print(list(filter(lambda x: x > 2, n)))
```

## Reduce

```python
n = [4, 3, 2,1]
#print 24
def mult(lst1):
  prod = lst1[0]
  for i in range(1, len(lst1)):
    prod *= lst1[i]
  return prod

print(mult(n))
```
