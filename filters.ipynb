# Filters

```python
#required library
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
%matplotlib inline
```

## Linear filter

### Blur filter (box)

#### Color image

```python
image = cv2.imread('/content/Dog_Breeds.jpg') #read image
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) #convert to HSV
```

```python
#Blur filter
kernel = np.ones((5,5),np.float32)/25
blur = cv2.filter2D(image,-1,kernel) #cv2.blur()
```

```python
#Visualization
plt.figure(figsize=(22,12))

plt.subplot(121)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_HSV2RGB))
plt.title('Original')

plt.subplot(122)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_HSV2RGB))
plt.title('Blur filter')
plt.show()
```

#### Graysclae image

```python
image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY) #convert to graysclae image
```

```python
#Box filter
kernel = np.ones((15,15),np.float32)/225 #np.ones((5,5),np.float32)/25
blur = cv2.filter2D(image2,-1,kernel) 
```

```python
#Visualization
plt.figure(figsize=(22,12))

plt.subplot(121)
plt.imshow(image2, cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(blur, cmap='gray')
plt.title('Blur filter')
plt.show()
```

### Gauss filter

```python
blur_gauss = cv2.GaussianBlur(image2,(9,9),21) #21
```

```python
#Visualization
plt.figure(figsize=(22,12))

plt.subplot(121)
plt.imshow(image2, cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(blur_gauss, cmap='gray')
plt.title('Gauss filter')
plt.show()
```

### Laplace filter

```python
edge_laplace = cv2.Laplacian(image2, 0, (5,5)) #24
```

```python
#Visualization
plt.figure(figsize=(22,12))

plt.subplot(121)
plt.imshow(image2, cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(edge_laplace, cmap='gray')
plt.title('Laplace filter')
plt.show()
```

### Sobel

```python
img = cv2.imread('/content/Dog_Breeds.jpg',0)

#https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_gradients/py_gradients.html
# Output dtype = cv2.CV_8U
sobelx8u = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
plt.figure(figsize=(22,22))

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

## Non-linear filter

### Median filter

```python
img = cv2.imread('/content/Dog_Breeds.jpg',0)
```

```python
blur_median = cv2.medianBlur(img, 9)
```

```python
plt.figure(figsize=(22,6))

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(blur_median, cmap='gray')
plt.title('Median filter')
plt.show()
```
